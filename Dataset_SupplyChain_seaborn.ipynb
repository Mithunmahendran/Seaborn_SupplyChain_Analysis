```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path='/content/supply_chain_data.csv'
df=pd.read_csv(path)
df
```

```python
ax=sns.barplot(x='Product type',y='Number of products sold',data=df,palette='magma',errorbar=None,)
plt.title('Analysis on quantites sold segmented by product types')
plt.xlabel('Product segment')
plt.ylabel('Units Sold')
for i in ax.containers:
  ax.bar_label(i)
```

The above barplot illustrates the top sold product category. From the plot it is evident that the market demand is high for skin care segment followed by cosmetics and then hair care.

```python
ax=sns.barplot(x='Product type', y='Revenue generated',data=df,palette='rocket',errorbar=None)
plt.title('Revenue analysis segmented by product type')
plt.xlabel('Product segment')
plt.ylabel('Revenue generated')
for i in ax.containers:
  ax.bar_label(i)
```

The barplot shows the analysis of the top revenue generated segment. The analysis concludes that even though the market demand is high for skin care segment from the previous analysis, the top revenue is generated by cosmetics, followed by skincare and then hair care.

```python
plt.figure(figsize=(8,7))
ax=sns.barplot(y='Order quantities',x='Product type',data=df,palette='cubehelix',errorbar=None)
plt.title('Order quantities based on product type')
for i in ax.containers:
  ax.bar_label(i)
```

As analysed previously the top order quantity is for the skincare segment, followed by cosmetics and then haircare.

```python
plt.hist(df['Shipping times'],color='Yellow',edgecolor='black')
plt.title('Shipping Time Distribution')
plt.xlabel('Shipping time')
plt.ylabel('Frequency')
```

The above histogram analyses the most frequent shipping time. From the plot, we can conclude that shipping time 8 is the most frequently used time.

```python
ax=sns.countplot(x='Shipping times',data=df,palette='magma')
plt.title('Analysis on shipping time frequency')
plt.xlabel('Shipping time')
plt.ylabel('frequency')
for i in ax.containers:
  ax.bar_label(i)
```

The same analysis on shipping time has been carried out on a countplot. Again tha analysis concludes that time 8 is the most frequently used one.

```python
plt.figure(figsize=(13,10))
ax=sns.barplot(x='Location',y='Shipping costs',data =df,hue='Shipping carriers',palette='YlOrBr',errorbar=None)
plt.title('Analysis of region wise shipping cost against each carrier')
plt.xlabel('Location')
plt.ylabel('Shipping Cost')
for i in ax.containers:
  ax.bar_label(i)
```

The bar plot analysis reveals the distribution of shipping costs across different regions, segmented by carriers. From the graph it is evident that carrier B has the highest shipping cost for Mumabi Kolkata and Delhi and carrier C for Banglore and carrier A Chennai

```python
Carrier_cost=df.groupby('Shipping carriers')['Shipping costs'].sum().sort_values(ascending=False).reset_index()
Carrier_cost
```

```python
ax=sns.barplot(x='Shipping carriers',y='Shipping costs',data=Carrier_cost,hue='Shipping carriers',errorbar=None)
plt.title('Analysis of total shipping cost distribution against each shipping carriers')
for i in ax.containers:
  ax.bar_label(i)
```

The above barplot analyses the distribution of total shipping cost against each shipping carrier. From the plot, the data reveals that Carrier B has the highest shipping cost followed by Carrier C and then Carrier A

```python
Shipping_costs = df.groupby('Supplier name')['Shipping costs'].sum().sort_values(ascending=False).reset_index()
```

```python
Shipping_costs
```

```python
ax=sns.barplot(x='Supplier name',y='Shipping costs',data=Shipping_costs,hue='Supplier name',errorbar=None)
plt.title('Analysis on shipping cost against each supplier')
for i in ax.containers:
  ax.bar_label(i)
```

The above bar graph analyses the distribution of shipping costs against each supplier. The data indicates that Supplier 1 has incurred the highest shipping cost followed by Supplier 2, Supplier 5, Supplier 4 and then supplier 3.

```python
ra1=df.groupby('Supplier name')['Shipping costs'].sum().sort_values(ascending=True).reset_index()
ra1
```

```python
ra2=df.groupby('Supplier name')['Revenue generated'].sum().sort_values(ascending=True).reset_index()
ra2
```

```python
merged_df=pd.merge(ra1,ra2, on='Supplier name')
merged_df
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x='Supplier name',y='Shipping costs',data=merged_df,marker='o',label='Shipping cost')
sns.lineplot(x='Supplier name',y='Revenue generated',data=merged_df,marker='o',label='Revenue generated')
plt.title('Shipping cost and Revenue generated by supplier')
plt.xlabel('Supplier name')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.legend()
```

The above lineplot illustrates the comparison between shipping cost and revenue generated by each supplier. From the plot it is clear that Supplier 1 has generated more revenue with lowe shipping cost followed by supplier 2, supplier 5 supplier 3 and then supplier 4

```python
plt.figure(figsize=(10,7))
ax=sns.barplot(x='Product type',y='Shipping times',estimator=np.mean,data=df,hue='Transportation modes',palette='viridis',errorbar=None)
plt.title('Analysis on average shipping time based on transport modes for each product type')
plt.xlabel('Product type')
plt.ylabel('Avg Shipping time')
for i in ax.containers:
  ax.bar_label(i)
```

The above barplot illustrates the average shipping time for each product type over various transportation modes.

```python
plt.figure(figsize=(9,9))
ax=sns.barplot(y='Number of products sold',x='Location',data=df,hue='Product type',palette='viridis',errorbar=None)
plt.title('Sales analysis segmented by region')
plt.xlabel('Location')
plt.ylabel('Units sold')
for i in ax.containers:
  ax.bar_label(i)
```

The above barplot illustrates the sales done over each region segmented by product type. The barplot indicates that Delhi has seen the highest sales so far

```python
ax=sns.countplot(x='Product type', data=df,hue='Transportation modes',palette='YlOrBr')
plt.title('Defect rates for each product type based on transportation modes')
plt.xlabel('Product segment')
plt.ylabel('Defect rate')
for i in ax.containers:
  ax.bar_label(i)
```

The above barplot analysis shows the defect rate for each product type against various transportation modes. From the plot it is evident that Air transport has the highest defect rate for skincare products and for hair care it is road transport mode and for cosmetics products the highest defect rate is through rail transport.

```python
ax=sns.barplot(x='Product type',y='Defect rates',data=df,estimator=np.mean,palette='Spectral',errorbar=None)
plt.title('Average defect rate for each product segment - Including all modes of transport')
plt.xlabel('Product segment')
plt.ylabel('Defect rates')
for i in ax.containers:
  ax.bar_label(i)
```

The above analysis shows that combining all transport mode the average  defect rate is higher for hair care products followed by skincare and then cosmetics

```python
ax=sns.barplot(x='Product type',y='Lead time',hue='Transportation modes',data=df,errorbar=None,estimator=np.mean,palette='rocket')
plt.title('Average Lead time by transportation mode for each product segment')
plt.xlabel('Product Segment')
plt.ylabel('Lead time')
for i in ax.containers:
  ax.bar_label(i)
```

Above barplot displays the average lead time for each product segment based on the various transportation modes.
The analysis concludes that for hair care products the average lead time is higher while using Sea transport and lowest while using air transport.
For Skincare segment the lead time is higher while using Road and Rail transport and lowest while using sea transprot.
For Cosmetics, the lead time is higher while using rail transport and the lowest while using air transport.
